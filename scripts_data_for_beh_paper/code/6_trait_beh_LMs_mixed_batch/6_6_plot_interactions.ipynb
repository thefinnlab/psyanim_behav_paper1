```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
root_dir = '../../data/subtlety_playfight_data/'#/Users/f0053cz/Documents/psyanim_v2/scripts_for_beh_paper/data/subtlety_playfight_data/'
data = pd.read_csv(f'{root_dir}curvefit_traits_merged_linearpf_withPC1.csv',index_col=[0])
data.head()
```

```python
sns.regplot(data=data,x='range_subt',y='PC1')
```

```python
sns.regplot(data=data,x='bias_subt',y='PC1')
```

```python
df1 = data.sort_values(by=['bias_subt'])
df1
```

```python
sns.regplot(data=df1.iloc[:20,:],x='range_subt',y='PC1')
sns.regplot(data=df1.iloc[-20:,:],x='range_subt',y='PC1')
```

```python
df2 = data.sort_values(by=['sigma_subt'])
df2 = df2.loc[df2['PSE_pf']!=0,:]
```

```python
sns.regplot(data=df2.iloc[:20,:],x='PSE_pf',y='PC1',label = '20 subs with the lowest sigma_subt (quickest flips)')
sns.regplot(data=df2.iloc[-20:,:],x='PSE_pf',y='PC1',label = 'subs with the highest sigma_subt (slowest flips)')
plt.legend()
```

```python
from scipy.stats import zscore
```

```python
# Z-score normalization for all columns
df2 = df2.apply(zscore)
```

```python
sns.regplot(data=df2.iloc[:20,:],x='PSE_pf',y='PC1',label = '20 subs with the lowest sigma_subt (quickest flips)')
sns.regplot(data=df2.iloc[-20:,:],x='PSE_pf',y='PC1',label = 'subs with the highest sigma_subt (slowest flips)')
plt.legend()
```
