```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(f'../../data/subtlety_playfight_data/subtlety_data.csv',index_col=[0]) 
df
```

```python
df_mean = df.groupby(['subID','subtlety']).mean().reset_index()
df_mean.drop(columns=['trial_num','stim_dur','subt_norm','chargeSpeed_norm','chargeSpeeds','responses_pf_rev','chargeSpeed_norm_rev'],inplace=True)
df_mean['rating180'] = [df_mean.loc[(df_mean['subID']==subID)&(df_mean['subtlety']==180),'responses_subt_1'].values[0] for subID in df_mean['subID']]
df_mean['diff_from_180'] = df_mean['responses_subt_1']-df_mean['rating180']
df_mean
```

```python
df_mean.to_csv(f'../../data/subtlety_data/mean_scores_subtlety.csv')
```

```python
df_median = df.groupby(['subID','subtlety']).median().reset_index()
df_median.drop(columns=['trial_num','stim_dur','subt_norm','chargeSpeed_norm','chargeSpeeds','responses_pf_rev','chargeSpeed_norm_rev'],inplace=True)
df_median['rating180'] = [df_median.loc[(df_median['subID']==subID)&(df_median['subtlety']==180),'responses_subt_1'].values[0] for subID in df_mean['subID']]
df_median['diff_from_180'] = df_median['responses_subt_1']-df_median['rating180']
df_median.head(14)
```

```python
curvefit_traits=pd.read_csv(f'../../data/subtlety_playfight_data/curvefit_traits_merged_linearpf_onlyneccols.csv',index_col=[0])
sel_cols=['subID']
sel_cols.extend(curvefit_traits.columns[np.where(curvefit_traits.columns=='social_skill')[0][0]:])
curvefit_traits = curvefit_traits[sel_cols]
curvefit_traits
```

```python
df_merged_mean = df_mean.merge(curvefit_traits,left_on='subID',right_on='subID')
df_merged_mean
```

```python
df_merged_mean.to_csv(f'../../data/subtlety_playfight_data/mean_mixedbatch_traits.csv')
```

```python
df_merged_median = df_median.merge(curvefit_traits,left_on='subID',right_on='subID')
df_merged_median
```

```python
df_merged_median.to_csv(f'../../data/subtlety_playfight_data/median_mixedbatch_traits.csv')
```
