```python
import pandas as pd
import numpy as np
```

```python
sesID=2
```

```python
df = pd.read_csv(f'../../data/subtlety_data/all_apr_ses{sesID}.csv') 
df
```

```python
df_mean = df.groupby(['subID','subtlety']).mean().reset_index()
df_mean['rating180'] = [df_mean.loc[(df_mean['subID']==subID)&(df_mean['subtlety']==180),'slider'].values[0] for subID in df_mean['subID']]
df_mean['diff_from_180'] = df_mean['slider']-df_mean['rating180']
df_mean
```

```python
df_mean.to_csv(f'../../data/subtlety_data/mean_apr_ses{sesID}.csv')
```

```python
df_median = df.groupby(['subID','subtlety']).median().reset_index()
df_median['rating180'] = [df_median.loc[(df_median['subID']==subID)&(df_median['subtlety']==180),'slider'].values[0] for subID in df_median['subID']]
df_median['diff_from_180'] = df_median['slider']-df_median['rating180']
df_median
```

```python
df_median.to_csv(f'../../data/subtlety_data/median_apr_ses{sesID}.csv')
```

```python
curvefit_traits=pd.read_csv(f'../../data/subtlety_data/curve_trait_apr_ses{sesID}.csv')
sel_cols=['subID']
sel_cols.extend(curvefit_traits.columns[np.where(curvefit_traits.columns=='aq_social_skill')[0][0]:])
curvefit_traits = curvefit_traits[sel_cols]
curvefit_traits
```

```python
df_merged_mean = df_mean.merge(curvefit_traits,left_on='subID',right_on='subID')
df_merged_mean
```

```python
df_merged_mean.to_csv(f'../../data/subtlety_data/mean_apr_ses{sesID}_traits.csv')
```

```python
df_merged_median = df_median.merge(curvefit_traits,left_on='subID',right_on='subID')
df_merged_median
```

```python
df_merged_median.to_csv(f'../../data/subtlety_data/median_apr_ses{sesID}_traits.csv')
```

```python
df_merged_alltrials = df.merge(curvefit_traits,left_on='subID',right_on='subID')
df_merged_alltrials
```

```python
df_merged_alltrials.to_csv(f'../../data/subtlety_data/mean_apr_ses{sesID}_traits_alltrials.csv')
```
